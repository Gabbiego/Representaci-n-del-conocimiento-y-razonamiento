```python
# HECHOS

hechos = {
    "es_empresa": ["EmpresaX"],
    "trata_datos_personales": ["EmpresaX"],
    "realiza_transferencia_internacional": ["EmpresaX"],
    "tiene_aviso_privacidad": [],
    "tiene_consentimiento_expreso": [],
    "incumple_LFPDPPP": [],
    "riesgo_sancion": [],
    "nivel_riesgo": {}
}

# REGLA 1
def regla_aviso(hechos):
    nuevos = []
    for e in hechos["es_empresa"]:
        if (e in hechos["trata_datos_personales"] and
            e not in hechos["tiene_aviso_privacidad"]):
            nuevos.append(e)
    return nuevos

# REGLA 2
def regla_transferencia(hechos):
    nuevos = []
    for e in hechos["es_empresa"]:
        if (e in hechos["realiza_transferencia_internacional"] and
            e not in hechos["tiene_consentimiento_expreso"]):
            nuevos.append(e)
    return nuevos

# REGLA 3
def regla_riesgo(hechos):
    return hechos["incumple_LFPDPPP"]

# REGLA 4
def regla_nivel(hechos):
    nuevos = {}
    for e in hechos["riesgo_sancion"]:
        nuevos[e] = "alto"
    return nuevos

# MOTOR DE INFERENCIA
def motor_inferencia(hechos):

    incumplimientos = regla_aviso(hechos)
    incumplimientos += regla_transferencia(hechos)

    hechos["incumple_LFPDPPP"] = list(set(incumplimientos))

    hechos["riesgo_sancion"] = regla_riesgo(hechos)

    hechos["nivel_riesgo"] = regla_nivel(hechos)

    return hechos

# EJECUCIÓN
resultado = motor_inferencia(hechos)
print(resultado)
```
